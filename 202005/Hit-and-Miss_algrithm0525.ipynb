# Approximation of Pi Using Hit and Miss (Monte Carlo) Method

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10000

x = np.random.uniform(low=-1, high=1, size=[N, 1])
y = np.random.uniform(low=-1, high=1, size=[N, 1])

inside_bool = x**2+y**2 < 1

approx_pi = 4*np.sum(inside_bool)/N
print('Pi: {}, approximation : {}'.format(np.pi, approx_pi))

x_in = x[inside_bool]
y_in = y[inside_bool]

plt.figure(figsize=[5, 5])
plt.scatter(x, y, s=1) #s...20
plt.scatter(x_in, y_in, color='r', s=1) #inside
plt.show()
```
