```r
if (!require(UpSetR)) install.packages("UpSetR")
```

```r
library('UpSetR')
movies <- read.csv( system.file("extdata", "movies.csv", package = "UpSetR"), header=T, sep=";" )
```

```r
upset(movies, nsets = 6, number.angles = 30, point.size = 3.5, line.size = 2, 
      mainbar.y.label = "Genre Intersections", sets.x.label = "Movies Per Genre", 
      text.scale = c(1.3, 1.3, 1, 1, 2, 0.75))
```

```r
library(UpSetR)

prot1 <- c("3U5Z", "3PT6", "1MHT", "1SSO", "1H1J", "1RIY")
prot2 <- c("3U5Z", "3PT6", "1MHT", "1T2K")
prot3 <- c("1SSO", "1H1J", "2KI2")
data <- list(A = prot1, B = prot2, C = prot3)

upset(fromList(data), order.by = "freq")
```

```r
sample_item <- read.csv(system.file("extdata", "sample_item.csv", package = "UpSetR"), header = T, sep = ";")
upset(sample_item, sets = c("TF1", "TF2", "TF3", "TF4"), order.by="degree")
```

```r
library(dplyr)
library(RColorBrewer)
library(UpSetR)

movies <- read.csv(system.file("extdata", "movies.csv",
                   package = "UpSetR"), header=T, sep=";" )
movies <- select(movies, Action:Children)

upset(movies, main.bar.color=brewer.pal(2^ncol(movies)-1, "Set1"))
```
