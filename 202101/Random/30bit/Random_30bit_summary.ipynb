```python
time_list = []
for num in [0, 1, 2, 3, 4]:
    df_path = '../../input/ost{}_{}.csv'.format(bit, num)
    df = pd.read_csv(df_path, sep=',')
    loop = 10
    result = timeit.timeit('ret.find_valid_y(df, num_reads)', globals=globals(), number=loop)
    time_list.append(result/loop)
    print('calucuration time of {}bit_{} data is {}'.format(bit, num, time_list[-1]))
print('mean calucuration time of {}bit data is {}'.format(bit, sum(time_list)/5))
```
