```python
X_mushroom = pd.DataFrame(index=np.arange(mushroom.shape[0]), columns=np.arange(max(mushroom[20])+1))
```

```python
X_mushroom.shape
```

```python
X_mushroom.head()
```

```python
for i in range(mushroom.shape[0]):
    for j in range(mushroom.shape[1]):
        X_mushroom.iloc[i, mushroom.iloc[i, j]] = 1
```

```python
X_mushroom.head()
```

```python
X_mushroom = X_mushroom.fillna(0)
```
