```python
import pandas as pd
import numpy as np
```

```python
mushroom = pd.read_csv("../input/mushroom.txt", sep=' ', header=None, index_col=None)
```

```python
mushroom.head()
```

```python
mushroom.shape
```

```python
#お試し小さめデータを作るために
while True:
    random_subset = mushroom.sample(n=5)
    y_s_mushroom = random_subset[21]
    if sum(y_s_mushroom)==2:
        y_s_mushroom.rename('y')
        y_s_mushroom.to_csv('../../202008/input/y_s_mushroom.csv')
        X_s_mushroom = pd.DataFrame(index=random_subset.index, columns=np.arange(max(random_subset[20])+1))
        
        for i in range(random_subset.shape[0]):
            for j in range(random_subset.shape[1]):
                X_s_mushroom.iloc[i, random_subset.iloc[i, j]] = 1
        
        X_s_mushroom = X_s_mushroom.fillna(0)
        X_s_mushroom.to_csv('../../202008/input/X_s_mushroom.csv')
        break
```

```python
while True:
    random_subset = mushroom.sample(n=16)
    y_l_mushroom = random_subset[21]
    if sum(y_l_mushroom)==8:
        y_l_mushroom.rename('y')
        y_l_mushroom.to_csv('../../202008/input/y_l_mushroom.csv')
        X_l_mushroom = pd.DataFrame(index=random_subset.index, columns=np.arange(max(random_subset[20])+1))
        
        for i in range(random_subset.shape[0]):
            for j in range(random_subset.shape[1]):
                X_l_mushroom.iloc[i, random_subset.iloc[i, j]] = 1
        
        X_l_mushroom = X_l_mushroom.fillna(0)
        X_l_mushroom.to_csv('../../202008/input/X_l_mushroom.csv')
        break
```
