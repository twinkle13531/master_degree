## Two Qubits

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
sigma_z1 = np.array([[1, 0], [0, -1]])
sigma_x1 = np.array([[0, 1], [1, 0]])
```

```python
d1 = 1
d2 = 1
```

```python
#H_P = 1/2*np.eye(2) + 1/2*sigma_z1 + 1/2*np.eye(2) + 1/2*sigma_z1
H_P = 1/2*np.eye(2) + 1/2*sigma_z1
H_B = d1*(1/2*np.eye(2) - 1/2*sigma_x1) + d2*(1/2*np.eye(2) - 1/2*sigma_x1)
```

```python
print(H_P)
print(H_B)
```

```python
eig_val_list = []
for s in [i / 10 for i in range(0, 11, 1)]:
    #print(s)
    Hchi_s = (1-s)*H_B + s*H_P
    #print(Hchi_s)
    eig_val, eig_vec =np.linalg.eig(Hchi_s) # 固有値・固有ベクトルをそれぞれeig_val, eig_vecに格納する
    print('eig_val', eig_val)
    #print('eig_vec', eig_vec)
    
    eig_val_list.append(eig_val)
```

```python
len(eig_val_list)
```

```python
plt.title("Fig1")
plt.xlabel("s") 
plt.ylabel("eigenvalues") 
plt.plot([i / 10 for i in range(0, 11, 1)], eig_val_list)
```

```python
eig_val_list = []
for s in [i / 10 for i in range(0, 11, 1)]:
    #print(s)
    Hchi_s = (1-s)*H_BB + s*H_P
    #print(Hchi_s)
    eig_val, eig_vec =np.linalg.eig(Hchi_s) # 固有値・固有ベクトルをそれぞれeig_val, eig_vecに格納する
    eig_val_list.append(eig_val)
```

```python
plt.title("Fig2")
plt.xlabel("s") 
plt.ylabel("eigenvalues") 
plt.plot([i / 10 for i in range(0, 11, 1)], eig_val_list)
```
