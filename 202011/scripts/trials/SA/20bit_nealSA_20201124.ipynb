## plot

### t1=9のときの、numreadsによるhit数変遷

```python
import matplotlib.pyplot as plt
```

```python
y= [72, 735, 2814]
x = [100, 1000, 10000]
```

```python
plt.xlabel('num_reads')
plt.ylabel('number of hits')
plt.plot(x, y)
plt.savefig('../../output/hit_numreads_9')
```

```python
y_x = [yy/xx for (yy, xx) in zip(y, x)]

plt.xlabel('num_reads')
plt.ylabel('number of hits / num_reads')
plt.plot(x, y_x)
plt.savefig('../../output/hit_numreads_9_persent')
```
