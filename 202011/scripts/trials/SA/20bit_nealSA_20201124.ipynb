## plot

```python
num_reads_list = [100, 1000, 10000]
num_reads_all_list = [nn*18 for nn in num_reads_list]
```

### t1=9のときの、numreadsによるhit数変遷

```python
import matplotlib.pyplot as plt
```

```python
y= [reads100_9, reads1000_9, reads10000_9]
```

```python
plt.xlabel('num_reads')
plt.ylabel('number of hits')
plt.plot(num_reads_list, y)
plt.savefig('../../../output/20bit_hit_numreads_9')
```

```python
y_x = [yy/xx for (yy, xx) in zip(y, num_reads_list)]

plt.xlabel('num_reads')
plt.ylabel('number of hits / num_reads')
plt.plot(num_reads_list, y_x)
plt.savefig('../../../output/20bit_hit_numreads_9_percentage')
```
