```python
def shisan_unyou(month_money, t):
    year_money = month_money * 12
    now_shisan = 0
    if t >= 1:
        for i in range(0,t):
            now_shisan = (year_money+now_shisan)*1.08
    else:
        now_shisan = year_money
    return now_shisan

def rieki(month_money, t):
    return shisan_unyou(month_money, t) - month_money*12*t

def rieki_nbai(month_money, t):
    nbai = shisan_unyou(month_money, t) / (month_money*12*t)
    return "あなたの資産は" + str(t) +"年後には、投資額に対して" +str(nbai) + "倍になる予定です"

def one_million_over(month_money, t):
    for t in range(50):
        if shisan_unyou(month_money, t) >= 10000:
            return "あなたの資産は" + str(t) + "年後には、1億円を超える予定です"
            break
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["font.size"] = 18 #x, y軸の文字サイズ
```

```python
month_money = 8 #月額
```

```python
t = np.arange(0, 30)
y = [[shisan_unyou(month_money, t)] for t in range(0, 30)]
fig, ax = plt.subplots(figsize=(10, 30)) #図のサイズ
ax.plot(t,np.array(y)) 
ax.set_yticks(np.linspace(0, 12000, 101)) #12000を最大値として100分割した
ax.set_xticks(np.linspace(0, 30, 16)) #30を最大値として15分割した
plt.grid(b=None, which='major', axis='both')
plt.xlabel("year")
plt.ylabel("assets")
plt.title("asset management")
plt.show()
```
